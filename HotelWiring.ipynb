```python
t = int(input("Number of test cases:"))
for i in range(t):
    m = int(input("Number of floors:"))
    n = int(input("Number of rooms per floor:"))
    k = int(input("number of switches to be turned off:"))
    
    # Four blank lists to store data
    list_t1 = []
    list_t2 = []
    list_t3 = []
    list_j = []
    
    for j in range(m):
        p = int(input("Number of rooms wired correctly:"))
        # Store the numbers of rooms wired correctly accordingly in the first list
        list_t1.append(p)
        # Store the index of each floor in the list that look similar to the indexing of each floor
        list_j.append(j)
        # Get the numbers of incorrect wiring per floor
        not_p = n - p
        # store the numbers of incorrect wiring in the third list
        list_t2.append(not_p)
        
    # Now put the floors index and the number of correct wiring per floor in a dictionary
    a = dict(zip(list_j, list_t1))
    # Now put the floors index and the number of incorrect wiring per floor in a dictionary
    c = dict(zip(list_j, list_t2))
    
    # This tests for various differences in the values of m and k
    if m - k == 0:
        # Provides a solution when all switches have been turned on
        total = 0
        b = 0
        while b < k:
            total += c.get(b)
            b += 1
        print(total)
    elif m - k == m:
        # Provides a solution when there is no switch turned off
        total = 0
        b = 0
        while b < k:
            total += a.get(b)
            b += 1
        print(total)
    else:
        # Lastly, provides a solution for various values in the number of switches turned off
        list_t1.sort()
        x = 0
        while x < k:
            y = list_t1[x]
            z = n - y
            list_t3.append(z)
            x += 1
        list_t3.extend(list_t1[k:])
        print(sum(list_t3))
```
