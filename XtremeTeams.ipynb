```python
t = int(input("No of test cases:"))
import math
i = 0
while i < t:
    n = int(input("No of students:"))
    m = int(input("No of topics:"))
    store1 = []
    
    # Getting the specialty of each student and making a check
    j = 0
    while j < n:
        specialty = input("Specialty of each student:")
        if len(specialty) != m:
            print("All courses weren't considered or limit was exceeded")
            break
        store1.append(specialty)
        j += 1
        
    # Checking if all the students have been considered
    if len(store1) != n:
        print("Restart please! Make sure all the courses are considered")
        break
    
    
    def split_str(item1):
        """Takes in a string as input and makes each element of the string
        an element of a new list"""
        new_list = []
        for element1 in item1:
            new_list.append(element1)
        return new_list
    
    # Split each item of store1 for easy indexing
    store2 = []
    for element2 in store1:
        store2.append(split_str(element2))
    
    
    def check_list(item2, position):
        """Takes in a list containig several lists alongside an integer.
        Checks for the numbers of y in the in that position of each list in the bigger list"""
        ans = 0
        for element2 in item2:
            if element2[position] == 'y':
                ans += 1
        return ans
    
    
    def count_val(item3, number):
        """Takes ina list of integers and counts the number of times the
        other input appears"""
        ans3 = 0
        for number in item3:
            if number == 1:
                ans3 += 1
        return ans3
    
    
    # Now check how difficult each topic is based on number of student specialized in the topic
    k = 0
    ans1 = 0
    store3 = []
    while k < m:
        store3.append(check_list(store2, k))
        k += 1
    store3.sort()
    
    # Result
    if 0 in store3 or count_val(store3, 1) == len(store3):
        # It will be impossible to choose a team since there's no student specialized in one or more of the topics
        print(0)
    elif count_val(store3, 1) < len(store3):
        total = count_val(store3, 1)
        left = 3 - total
        left_to_pick = n - total
        print(math.comb(left_to_pick, left))
    else:
        print(math.comb(n,3))
        
    i += 1
        
```
